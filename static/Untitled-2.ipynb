```python
import matplotlib.pyplot as plt
import numpy as np

# Initialize drone positions
num_drones = 5
positions = np.random.rand(num_drones, 2) * 10  # Random positions

plt.ion()  # Enable interactive mode
fig, ax = plt.subplots()
sc = ax.scatter(positions[:, 0], positions[:, 1], c='red', label="Drones")
ax.set_xlim(0, 10)
ax.set_ylim(0, 10)
ax.legend()

# Simulate movement
for _ in range(50):  # 50 time steps
    positions += (np.random.rand(num_drones, 2) - 0.5)  # Random movement
    sc.set_offsets(positions)
    plt.pause(0.1)  # Pause for animation effect

plt.ioff()  
plt.show()
```
