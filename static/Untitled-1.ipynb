```python
import sklearn
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn import svm
import skimage
from skimage import io
import os
import numpy as np
import matplotlib.pyplot as plt
import pandas as  pd
from sklearn import preprocessing
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import Normalizer
from skimage.io import imread
from skimage.color import rgb2gray
import joblib
from skimage.transform import resize
```

```python
from sklearn.svm import SVC
```

```python
input_path ='C:/Users/Victor/OneDrive/Documents/hackthorn/dataset'

cartegories =['organic','recyclable']
```

```python
data = []
labels = []
limit_per_category = 400  # Limit to 100 images per category

# Process images
for category_index, category in enumerate(cartegories):
    category_path = os.path.join(input_path, category)

    if not os.path.exists(category_path):
        print(f"Warning: {category_path} does not exist!")
        continue

    files = os.listdir(category_path)[:limit_per_category]  # Limit file processing to 100

    for file in files:
        img_path = os.path.join(category_path, file)
        print(f"Processing: {img_path}")

        try:
            image = imread(img_path).astype(np.float32)  # Read image as float32
            if len(image.shape) == 3:  # If RGB, convert to grayscale
                image = resize(image, (256, 256))
                image = rgb2gray(image)

            data.append(image.flatten())
            labels.append(category_index)
        except Exception as e:
            print(f"Error processing {img_path}: {e}")

# Convert to NumPy arrays
data = np.array(data)
labels = np.array(labels)

print(f"Processed {len(data)} images with {len(set(labels))} categories.")
```

```python
X_train, X_test, y_train, y_test = train_test_split(data, labels, test_size=0.2,shuffle=True,stratify=labels)
```

```python
classifer=SVC(probability=True)
classifer.fit(X_train,y_train)
```

```python
y_pred=classifer.predict(X_test)
score=classifer.score(X_test,y_test)
print(score)
```
